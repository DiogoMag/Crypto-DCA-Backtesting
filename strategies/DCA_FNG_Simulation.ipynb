```python
import pandas as pd

# 1. Load data
pd.set_option('display.max_rows', None)
df = pd.read_csv('../data/BTC_FNG_merged.csv')

df.head(1)
```

```python
# Initialize
dca_amount = 100  # Buy $100 worth every 7 days
total_invested = 0
holdings = 0

portfolio_value = []
invested_amount = []

def calculate_signal(df, i):
    # Buy every 7 days
    if i % 7 == 0:
        return 'BUY'
    else:
        return 'HOLD'
     
# 3. Loop through data
for i in range(len(df)):
    price = df['close'].iloc[i]
    fng = 
    signal = calculate_signal(df, i)
    
    if signal == 'BUY':
        shares = dca_amount / price
        holdings += shares
        total_invested += dca_amount
    
    current_value = holdings * price
    portfolio_value.append(current_value)
    invested_amount.append(total_invested)

# Print results
print(f"Total invested: ${total_invested:.2f}")
print(f"Final portfolio value: ${portfolio_value[-1]:.2f}")
print(f"Gain/Loss: ${(portfolio_value[-1] - total_invested):.2f}")
print(f"ROI: {((portfolio_value[-1] - total_invested) / total_invested * 100):.2f}%")
```
