```python
import pandas as pd
```

```python
# 1. Load data
df = pd.read_csv('../data/BTC_FNG_merged.csv')
```

```python
# Define starting capital
starting_capital = 14700
```

```python
# 2. Initialize
portfolio_value = [starting_capital]
holdings = 0
cash = starting_capital
```

```python
def calculate_signal(df, i):
    # Buy every 7 days
    if i % 7 == 0:
        return 'BUY'
    else:
        return 'HOLD'
     
```

```python
# 3. Loop through data

dca_amount = 100  # Buy $100 worth every 7 days

for i in range(len(df)):
    price = df['close'].iloc[i]
    signal = calculate_signal(df, i)
    
    if signal == 'BUY' and cash >= dca_amount:
        shares = dca_amount / price
        holdings += shares
        cash -= dca_amount  # Subtract only the DCA amount, not all cash
    
    current_value = cash + (holdings * price)
    portfolio_value.append(current_value)

# Print results
print(f"Final portfolio value: ${portfolio_value[-1]:.2f}")
print(f"ROI: {((portfolio_value[-1] - starting_capital) / starting_capital * 100):.2f}%")
```
